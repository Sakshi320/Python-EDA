# Read this... then performed string substition (for $ , + and any other special characters) and convert `Per capitaincome`,	`Medianhouseholdincome`,	and `Medianfamilyincome` to numeric
* https://pandas.pydata.org/docs/reference/api/pandas.Series.str.replace.html
* https://stackoverflow.com/questions/43096522/remove-dollar-sign-from-entire-python-pandas-dataframe
* https://www.skytowner.com/explore/removing_comma_from_column_values_in_pandas_dataframe


```python
# ignore the warning! it's just a warning for a future dependency
#Converting Per Capitaincome to numeric-
df['Per capitaincome'] = df['Per capitaincome'].str.replace('$', '')
df['Per capitaincome'] = df['Per capitaincome'].str.replace(',', '')
df['Per capitaincome'] = pd.to_numeric(df['Per capitaincome'], errors='coerce')
```

```python
# ignore the warning! it's just a warning for a future dependency
#Converting Medianhouseholdincome to numeric-
df['Medianhouseholdincome'] = df['Medianhouseholdincome'].str.replace('$', '')
df['Medianhouseholdincome'] = df['Medianhouseholdincome'].str.replace(',', '')
df['Medianhouseholdincome'] = pd.to_numeric(df['Medianhouseholdincome'], errors='coerce')

#Converting Medianfamilyincome to numeric-
df['Medianfamilyincome'] = df['Medianfamilyincome'].str.replace('$', '')
df['Medianfamilyincome'] = df['Medianfamilyincome'].str.replace(',', '')
df['Medianfamilyincome'] = df['Medianfamilyincome'].str.replace('+', '')
df['Medianfamilyincome'] = pd.to_numeric(df['Medianfamilyincome'], errors='coerce')

df
```

# County has the highest average df['Rank']? The lowest `df['Rank]'

```python
print("Lowest rank of County is:",df.groupby(["County"]).mean()['Rank'].sort_values().head(1))
#sorted the values from low to high then displayed first/lowest value with head

print("Highest rank of County is:",df.groupby(["County"]).mean()['Rank'].sort_values().tail(1))
#sorted the values from low to high then displayed last/highest irst value with tail
```
